```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import random

# 1. Tạo dữ liệu giả lập
np.random.seed(42)  # Đảm bảo kết quả có thể tái tạo
items = ['Gạo', 'Đậu', 'Cá', 'Rau', 'Thịt', 'Nước mắm']  # Danh sách sản phẩm
num_transactions = 1000  # Số giao dịch

# Tạo 1000 giao dịch ngẫu nhiên
transactions = []
for _ in range(num_transactions):
    # Mỗi giao dịch chứa từ 1 đến 4 sản phẩm ngẫu nhiên
    transaction_size = random.randint(1, 4)
    transaction = random.sample(items, transaction_size)
    transactions.append(transaction)
print("Danh sách giao dịch:")
print(transactions)
# 2. Chuyển dữ liệu sang định dạng one-hot encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

print("\nDataFrame one-hot encoding:")
print(df)
# 3. Chạy thuật toán Apriori
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)
print("Tập mục phổ biến:")
print(frequent_itemsets)

# 4. Sinh luật kết hợp
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1)
print("\nLuật kết hợp:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

# 5. Lưu kết quả vào file CSV (tùy chọn)
rules.to_csv('association_rules_output.csv', index=False)
print("\nKết quả luật kết hợp đã được lưu vào file 'association_rules_output.csv'")
```
